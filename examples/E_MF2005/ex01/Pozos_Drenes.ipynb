```python
import numpy as np
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D
#from matplotlib import cm
plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True
```

```python
def visMalla(nx, ny, x, y):
    for yv in y:
        ym = np.zeros(nx)
        ym[:] = yv
        plt.plot(x, ym, color='k',ls='-', lw=1.5)
    for xv in x:
        xm = np.zeros(ny)
        xm[:] = xv
        plt.plot(xm, y, color='k',ls='-', lw=1.5)
```

```python
Lx = 75000
Ly = 75000
ztop = 200
zbot = -450
Lz = ztop - zbot
wells = [( 6,12), ( 9, 8), ( 9,10), ( 9,12), ( 9,14), 
         (11, 8), (11,10), (11,12), (11,14), (13, 8), 
         (13,10), (13,12), (13,14),
         ( 4, 6), ( 5,11)]

nx = 15
ny = 15
nz = 5

dx = Lx / nx
dy = Ly / ny

x = np.linspace(0, Lx, nx+1)
y = np.linspace(0, Ly, ny+1)
xg, yg = np.meshgrid(x,y)

xc = np.arange(-dx*0.5, Lx+dx, dx)
yc = np.arange(-dy*0.5, Ly+dy, dy)

plt.figure(figsize=(20,6))

plt.subplot(131)
visMalla(nx+1,ny+1,x,y)
for c in wells[:-2]:
    plt.scatter(xc[c[1]], yc[c[0]], color='blue')

plt.xlim(-dx, 80000)
plt.ylim(80000, -dy)
plt.xticks(np.arange(0,Lx+1,15000))
plt.yticks(np.arange(0,Ly+1,15000))
plt.gca().set_aspect('equal')
plt.title('Layer 1', fontsize=20)
plt.ylabel('Rows')
plt.xlabel('Columns')

plt.subplot(132)
visMalla(nx+1,ny+1,x,y)
plt.scatter(xc[6], yc[4], color='red')
plt.xlim(-dx, 80000)
plt.ylim(80000, -dy)
plt.xticks(np.arange(0,Lx+1,15000))
plt.yticks(np.arange(0,Ly+1,15000))
plt.gca().set_aspect('equal')
plt.title('Layer 2', fontsize=20)
plt.ylabel(' ')
plt.xlabel('Columns')

plt.subplot(133)
visMalla(nx+1,ny+1,x,y)
plt.scatter(xc[11], yc[5], color='orange')
plt.xlim(-dx, 80000)
plt.ylim(80000, -dy)
plt.xticks(np.arange(0,Lx+1,15000))
plt.yticks(np.arange(0,Ly+1,15000))
plt.gca().set_aspect('equal')
plt.title('Layer 3', fontsize=20)
plt.ylabel(' ')
plt.xlabel('Columns')

plt.show()
```

```python
plt.figure(figsize=(10,8))
visMalla(nx+1,ny+1,x,y)
eti=True
for c in wells[:-2]:
    if eti:
        plt.scatter(xc[c[1]], yc[c[0]], color='blue', label='Layer 1')
        eti = False
    else:
        plt.scatter(xc[c[1]], yc[c[0]], color='blue')

plt.scatter(xc[6], yc[4], color='red', label='Layer 2')
plt.scatter(xc[11], yc[5], color='orange', label='Layer 3')

plt.xlim(-dx, 80000)
plt.ylim(80000, -dy)
plt.xticks(np.arange(0,Lx+1,15000))
plt.yticks(np.arange(0,Ly+1,15000))
plt.gca().set_aspect('equal')
plt.title('Wells', fontsize=20)
plt.ylabel('Rows')
plt.xlabel('Columns')
plt.legend(bbox_to_anchor=(1.30, 1), loc='upper right', ncol=1)

plt.savefig('Wells.png')
```

```python
drains = [(8,2),(8,3),(8,4),(8,5),(8,6),(8,7),(8,8),(8,9),(8,10)]

plt.figure(figsize=(8,8))
visMalla(nx+1,ny+1,x,y)
for c in drains:
    plt.scatter(xc[c[1]], yc[c[0]], color='green')
plt.xlim(-dx, 80000)
plt.ylim(80000, -dy)
plt.xticks(np.arange(0,Lx+1,15000))
plt.yticks(np.arange(0,Ly+1,15000))
plt.gca().set_aspect('equal')
plt.title('Drains - Layer 1', fontsize=20)
plt.ylabel('Rows')
plt.xlabel('Columns')
#plt.legend(bbox_to_anchor=(1.30, 1), loc='upper right', ncol=1)

plt.savefig('drains.png')
```
