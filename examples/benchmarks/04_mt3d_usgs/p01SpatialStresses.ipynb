```python
import os
import sys
from tempfile import TemporaryDirectory

import matplotlib as mpl
import numpy as np
import pandas as pd

import flopy

#! mf6.exe p01SpatialStresses.nam

print(sys.version)
print(f"numpy version: {np.__version__}")
print(f"matplotlib version: {mpl.__version__}")
print(f"flopy version: {flopy.__version__}")
```

```python
# temporary directory
temp_dir = TemporaryDirectory()
#model_ws = temp_dir.name
#model_ws = './'
model_ws = '.\\'
modelpth = os.path.join(model_ws, "p01SpatialStresses_4mf6_mt")
print(modelpth)
modelname = "p01SpatialStresses_4mf6_mt"
```

```python
# set up simulation and basic packages

#sim = flopy.mf6.MFSimulation(sim_name=name, sim_ws=modelpth)

sim = flopy.mf6.MFSimulation(sim_name="p01SpatialStresses", sim_ws=modelpth, exe_name="mf6.exe")

print("Done creating simulation.")
```

```python
# Make sure modelpth directory exists
if not os.path.isdir(modelpth):
    os.makedirs(modelpth, exist_ok=True)
```

```python
# Start by setting up MT3D-USGS class and use files created with MF6-GWF for setting up a number of the BTN arrays

mt = flopy.mt3d.Mt3dms(
    modelname=modelname,
    model_ws=modelpth,
    version="mt3d-usgs",
    namefile_ext="nam",
    exe_name="mt3d-usgs_1.1.0_64.exe",
    ftlfree=True,
)
```

```python
#sim.run_simulation(report=True)
```

```python
sim.run_simulation(report=True)
```
